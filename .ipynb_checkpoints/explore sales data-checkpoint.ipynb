## let's use quartiles

```python
Q20 = np.percentile(df_area['oms'], 20)
Q80 = np.percentile(df_area['oms'], 80)
print Q33,Q66
 
def perf2cat(x):
    """
    create a category for a given day 
    """
    if x < Q20:
        return "bad day"
    elif x < Q80:
        return "normal day"
    else:
        return "good day"
df_area['revenue'] = df_area['oms'].map(perf2cat)
```

### Build a simple regression model

```python
from sklearn.ensemble import RandomForestRegressor
from utils import oneHot
import random

features = ['month','precipitations','sun','temp','wind','weekday','station']
data = aggregated_data[features]
data = oneHot(data, 'weekday')
data = oneHot(data, 'station')
data = oneHot(data, 'month')
train_features = data.columns.values
labels = ['oms']
X = data.as_matrix()
y = aggregated_data[labels].values
data[:3]
```

```python
full_training = False
train_ratio = 0.9
L = len(y)
idx = range(L)
random.shuffle(idx)
if not full_training:
    X_train = X[ idx[ : int(0.8 * L)  ]  , : ]
    y_train = y[ idx[ : int(0.8 * L)  ] , : ]
else:
    X_train = X
    y_train = y
X_test = X[ idx[ int(0.8 * L)  : ]  , : ]
y_test = y[ idx[ int(0.8 * L)  : ] , :]

print X_train.shape
```

```python
rfr = RandomForestRegressor(n_estimators=200, min_samples_split=5)
rfr.fit(X_train, y_train)
```

```python
def getErr(pred,true):
    return 1 - abs( 1 - pred/true )
    
pred = rfr.predict(X_test)
scores = []
for i in range(len(pred)):
    e = getErr(pred[i],y_test[i])
    scores.append(e)
print "accuracy: " + str(np.mean(scores))
sns.distplot(scores)
```

86% of accuracy, quite a good score. 84% of accuracy was obtain 

```python
pairs = []
for i in range(len(train_features)):
    pairs.append( (train_features[i], rfr.feature_importances_[i] ) )
for u in sorted(pairs , key= lambda x: x[1] , reverse = True):
    print str(u[1]*100)[:6] + '% \t' ,
    print u[0]
```

According to the model, after the station, the weekday is the most important feature, then the temperature, the sun, the wind and finally the level of precipitations.
